```python
co2_df = pd.read_csv("co2 emissions.csv")
co2_df.head()
```

```python
co2_df.dtypes
```

```python
co2_df = co2_df.rename(columns={'state-name': 'state'})
co2_df.head()
```

```python
temp_df = temp_df.drop(columns=['Unnamed: 0'])
```

```python
temp_yearly = temp_df.groupby(['year', 'state']).mean()
temp_yearly = temp_yearly.drop(columns=['month']).reset_index()
temp_yearly.head()
```

```python
temp_yearly.dtypes
```

```python
join_df = pd.merge(co2_df, temp_yearly, on=['year', 'state'])
join_df
```

```python
test_data = join_df.groupby(['year', 'state']).agg({
    'value': 'sum',
    'average_temp': 'mean',
    'centroid_lon': 'first',
    'centroid_lat': 'first'
}).reset_index()
```

```python
test_data.to_csv('test_data.csv', index=False)
```
