```python
import pandas as pd
import numpy as np
```

```python
temp_df = pd.read_csv("temperature.csv")
temp_df.head()
```

```python
co2_df = pd.read_csv("co2 emissions.csv")
co2_df.head()
```

```python
co2_df.dtypes
```

```python
co2_df = co2_df.rename(columns={'state-name': 'state'})
co2_df.head()
```

```python
temp_df = temp_df.drop(columns=['Unnamed: 0'])
```

```python
temp_yearly = temp_df.groupby(['year', 'state']).mean()
temp_yearly = temp_yearly.drop(columns=['month']).reset_index()
temp_yearly.head()
```

```python
temp_yearly.dtypes
```

```python
join_df = co2_df.join(temp_yearly, how='left', lsuffix='_co2', rsuffix='_temp')
join_df
```
